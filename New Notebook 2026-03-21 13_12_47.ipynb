```python
# Ingest customer CSV files from landing volume via Auto Loader
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

customer_path = "/Volumes/workspace/landing/customer/"
customer_df = (spark.readStream.format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")
    .load(customer_path)
)

# Display sample records and schema
customer_df.printSchema()
display(customer_df.head(10))
```

```python
# Ingest orders CSV files from landing volume via Auto Loader
orders_path = "/Volumes/workspace/landing/orders/"
orders_df = (spark.read.format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("header", "true")
    .load(orders_path)
)

# Display sample records and schema
orders_df.printSchema()
display(orders_df.head(10))
```
