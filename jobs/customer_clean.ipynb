```python
from pyspark.sql.functions import col, lower, trim

df = spark.table("ingest_schema.customer_raw")

customer_clean = (
    df.withColumn("name", trim(col("name")))
      .withColumn("email", lower(col("email")))
)

customer_clean.write.mode("overwrite").saveAsTable("ingest_schema.customer_clean")
```
