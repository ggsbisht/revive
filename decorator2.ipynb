```python
def write_to_delta(func):
    def wrapper():
        df=func()
        table_name=func.__name__
        print(f"table name is {0}",table_name)
        df.write.format("delta").mode("overwrite").saveAsTable(table_name)
        print("delta table created")

    return wrapper
```

```python
@write_to_delta
def orders_bronzetable():
    df=spark.read.format("csv").option("inferSchema","true").option("header","true")\
        .load("/Volumes/workspace/landing/landing/orders/orders1.csv")
    return df
```
